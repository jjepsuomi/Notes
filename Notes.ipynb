## --- MATHEMATICS ---


### Topology 
A topology $\tau$ is a set of subsets of some other set $X$ with the properties that $X\in \tau$ and $\emptyset \in\tau$. Also, the intersections and unions of the members in $\tau$ are also in $\tau$. 

### Topological space
Set $Y$ is called topological space if $\tau$ is a topology in $Y$.

### $\sigma$-algebra
A $\sigma$-algebra $\omega$ is a set of subsets of some other set $X$ with the properties that $X\in\omega$, and for each $A\in \omega$ also $A^c \in \omega$. Also, the unions of the members in $\omega$ are also in $\omega$. 

### Measure 
Measure (usually denoted by $m$ or $\mu$) is positive valued function which satisfies the properties of a well-defined measure (e.g. length, area, volume). You can think of measure $m$ representing the "size" of something, like for example a set of numbers. For example the length of interval $[0, 1]$ is 1, that is $m([0, 1]) = 1.$ Furthermore, measure must satisfy also other conditions such as countable-additivity (whole equals the sum of its parts) et cetera. 

### Measurable space
Set $X$ is called measurable space if $\omega$ is a $\sigma$-algebra in $X$.

### Measurable set 
Members of $\sigma$-algebra $\omega$ are called measurable sets.

### Measurable function
A function $f:X\to Y$, where $X$ is a measurable space and $Y$ a topological space, is called a measurable function if for every open set $V\in Y$ we have that $f^{-1}(V)\in \omega$, where $\omega$ is a $\sigma$-algebra in $X$ (thus $f^{-1}(V)$ is a measurable set). That is, we must be able to "measure" the preimages of $f$.



## --- CHEMISTRY ---

**Law of conservation of mass (Lavoisier)** = Mass cannot be created nor destroyed

**Law of definite proportion (Proust)** = Given compound always contains exactly same proportions of given elements by mass

**Law of multiple proportions (Dalton)** = When two elements form a series of
compounds, the ratios of the masses of the second element that combine with
1 gram of the first element can always be reduced to small whole numbers.

**Dalton’s atomic theory**:
■ All elements are composed of atoms.
■ All atoms of a given element are identical.
■ Chemical compounds are formed when atoms combine.

**Electrons in reactions** = The behavior of an atom in chemical reactions is determined by its electrons

**Atomic mass number** = # protons + # neutrons

**Atomic number** = # protons (denoted also as Z in $^A_Z X$). In uncharged atom, Z = number of electrons

**Covalent bond** = atoms sharing electrons

**Molecule** = Group of atoms

**Chemical formula** = representation of the numbers and types of atoms present 

**Structual formula** = Drawn representation of the structure of the atom

**Ion** = atom with net positive or negative charge

**Isotope** = atoms are isotops when they contain same number of protones but different number of neutrons

**Cation** = ion with net positive charge (e.g. $Na \rightarrow Na^{+} + e^-$)

**Anion** = ion with net negative charge (e.g. $Cl + e^- \rightarrow Cl^-$)

**polyatomic ion** = ion consisting from multiple atoms

**oxyanions** = polyatomic anions with differing number of oxygen atoms

**Ionic bond** = force of attraction between oppositely charged ions 

**Ionic solid (salt)** = A solid consisting of oppositely charged ions

**Metals** = atoms which tend to lose electorns, thus forming cations

**Non-metals** = atoms which tend to gain electorns, thus forming anions

**Periodic table:** elements in same column tend to have similar chemical properties

**Acid** = a molecule with one or more $H^+$ ions attached to an anion

**Naming conventions**:
- Type I binary ionic compound (monoatoms, cation-anion)-> "Cation element name" "Anion element name + ending in -ide". E.g. Sodium Chloride $NaCl$ (ions $Na^+, Cl^-$), Lithium Nitride $Li_3N$ (ions $Li^+, N^{3-}$), Magnesium oxide $MgO$ (ions $Mg^{2+}, O^{2-}$). Type I cations form only one type on cations.


- Type II binary ionic compund (monoatoms, cation-anion)-> Almost same naming as in Type I, but roman numeral needs to be specified for cation charge. Type II can have multiple different cations (number identifies). If only one cation for ion -> roman numeral not needed. E.g. $Fe^{2+}, Fe^{3+}$ = Iron(II), Iron(III), also older: Ferrous (smaller charge -ous), Ferric (larger charge -ric). Examples: $CoBr_2$ -> Cobalt(II) bromide (Cobalt has multiple cations, compund contains two $Br^-$ anions). $Al_2O_3$ Aluminum oxide, Aluminum forms only $Al^{3+}$ cation, no numeral needed.


- Oxyanions: different numbers of oxygen atoms (e.g. hypo- smallers, -ite, -ate, per-) $ClO^-, ClO^-_2, ClO^-_3, ClO^-_4$ -> Hypochlorite, Chlorite, Chlorate, Perchlorate.


- Type III binary covalent bond (two non-metals) -> Similar as in ionic bonds. "Element name normally" + "Second element treated like anion -ide". Prefixes used to tell numbers of atoms present e.g. mono-, di-, tri-, tetra-, penta-. E.g. Nitrogen Dioxide $NO_2$, Dinitrogen oxide $N_2O$, Dinitrogen pentoxide $N_2O_5$. "mono-" is never used in front of first element. Note that vowel -a sometimes left off to make sound better, e.g. pentoxide(YES) vs. pentaoxide(NO)

- Acid naming: depends on whether anion contains oxygen. If does NOT contain oxygen --> prefix: "hydro-" and suffix "-ic" (hydro-anion name-ic), e.g. $HCl, HCN, H_2S$ = hydrochloric, hydrocyanic and hydrosulfuric acid. If DOES contain oxygen --> "root name of anion" + suffix "-ic" or "-ous". If anion name ends in "-ate" acid name ends in "-ic" (or "-ric"). E.g. $H_2SO_4$ (sulfate anion), $HC_2H_3O_2$ (acetate ion) -> sulfuric acid, acetic acid. If anion has "-ite" ending acid name ends with "-ous". E.g. $H_2SO_3$ (sulfite anion) is sulfurous acid. 




**Relationship between EM wavelength, frequency and speed of light**: $\lambda f = c$

**Change in energy for a system $\Delta E$ can be represented by the equation**: $\Delta E = nhf$, where $n\in{1,2,3,...}$, $h$ is Planck's constant and $f$ is the frequency of EM radiation emitted or absorbed. Planck: gained or lost energy is quantized. Packets of energy are called "quantum", "quanta" (many). 

**Energy of photon (Einstein)**: $E_{\text{photon}} = hf$

**Excess kinetic energy of electron:** $KE_{\text{electron}} = \frac12mv^2 = hv-hv_0$, where $v,v_0$ is the photon frequency and minimum frequency needed to extract electron from surface. $m,v$ are the mass and speed of electron. 

**Relativity (Einstein)**: When system loses energy, it loses mass $m = \frac{E}{c^2}$

**Quantum mechanics**: matter has both particle and wave like beahvior. Particle with momentum $mv$ has wavelength $\lambda = \frac{h}{mv}$ 

**Atomic spectrum of atom:** only certain energies are allowed for the electron in atoms. That is, energy is quantized in atoms. Discrete spectra, not continuous (would be the case if all energy states were possible). High energy charge passed through gasses --> molecules breaks, exited with energy, excess energy emits radiation --> only certain energies emit radiation --> quantized energy. E.g. change from higher to lower energy level in atom (excess energy radiated). Wavelength can be solved from: $\Delta E = \frac{hc}{\lambda}$

**Bohr model:** electron in a (hydrogen) atom moves around the nucleus only in certain allowed circular orbits. Because electron revolves around nucleus, it is constantly accelerating, and should thus constantly lose energy and radiate EM waves.
